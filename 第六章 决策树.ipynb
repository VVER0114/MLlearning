# 6.1 决策树分类与绘制

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:,2:]
y = iris.target

tree_clf = DecisionTreeClassifier()
tree_clf.fit(X, y)
```

```python
from sklearn import tree
tree.plot_tree(tree_clf)
```
