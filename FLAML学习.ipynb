# FLAML - Fast and Lightweight AutoML
微软轻量级AutoML应用
支持分类、回归、时间序列

```python
from flaml import AutoML
from sklearn.datasets import load_boston
```

```python
from sklearn.model_selection import train_test_split
automl = AutoML()
X, y = load_boston(return_X_y=True)
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import r2_score,average_precision_score
automl.fit(X_train, y_train, task="regression", estimator_list=["lgbm"])
```

```python
y_pred = automl.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

## 参数设置
**两个*不能漏**

```python
automl_settings = {
    "time_budget": 10,  # in seconds
    "metric": 'r2',
    "task": 'regression',
    "log_file_name" : 'test.log'
}
```

 ## 参数调节
 1. **task**，训练目标，目前有‘classification’, ‘regression’, ‘ts_forecast’, ‘rank’.'LRL2Classifier'
 2. **metric**，训练分数，‘accuracy’, ‘roc_auc’, ‘roc_auc_ovr’, ‘roc_auc_ovo’, ‘f1’, ‘micro_f1’, ‘macro_f1’, ‘log_loss’, ‘mae’, ‘mse’, ‘r2’, ‘mape’. **可自定义**
 3. **estimator_list**，估计器方法，有['lgbm', 'xgboost', 'catboost', 'rf', 'extra_tree']，**加方括号**
 4. **n_jobs** 
 5. **time_budget** ，训练时间，单位s
 6. **max_iter**，迭代次数
 7. **early_stop** default=False | Whether to stop early if the search is considered to converge.
 8. **log_file_name** – A string of the log file name.

```python
automl.fit(X_train=X_train, y_train=y_train,
           **automl_settings)
```

## 分类问题

```python
from flaml import AutoML
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
automl = AutoML()
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
automl.fit(X_train, y_train, task="classification")
```

```python
y_pred = automl.predict(X_test)
```

```python
automl.model.estimator
```

```python
from flaml.ml import sklearn_metric_loss_score
print('accuracy', '=', sklearn_metric_loss_score('accuracy', y_pred, y_test))
print('f1_score', '=', sklearn_metric_loss_score('micro_f1', y_pred, y_test))
```
